```python
import plotly.graph_objects as go
import ipywidgets
import solara as sl

f = go.FigureWidget()
f
```

```python
import vaex as vx
df = vx.example()[:1_000].to_pandas()
df
```

```python
import plotly.express as px
fig = px.scatter(df, x='x',y='y')
fig.update_layout(
    height=800,
    autosize=True,
)
fig.show()
```

```python
import reacton.ipyvuetify as rv
@sl.component
def Page():
    df = vx.example()[:1_000].to_pandas()
    with rv.Card(style_=" width: 100%; height: 100%") as main:
        with rv.CardText():
            fig = px.scatter(df, x='x',y='y')
            sl.FigurePlotly(fig)
    return main
Page()
```
