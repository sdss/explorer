```python
import bokeh
from jupyter_bokeh import BokehModel
from bokeh.io import output_notebook, show

from bokeh.models import Plot, Scatter, ColumnDataSource, LinearAxis, LogAxis, LogTickFormatter, CategoricalTickFormatter, BasicTickFormatter, DataRange1d, LogTicker, Grid, BasicTicker, CompositeTicker
from bokeh.transform import linear_cmap, log_cmap, factor_cmap
from bokeh.models import (
    BoxSelectTool,
    BoxZoomTool,
    WheelZoomTool,
    LassoSelectTool,
    ColorBar,
    HoverTool,
    Scatter,
    ExamineTool,
    TapTool,
    PanTool,
    ResetTool,
    PrintfTickFormatter,
    CustomJSTickFormatter,
)

import vaex as vx
import numpy as np

df = vx.example()[:10_000]
    

cds = ColumnDataSource(data={'x': np.log10(df['x'].values),'y': df['y'].values, 'z': df['z'].values})
p = Plot(x_range=DataRange1d(), y_range=DataRange1d(), width_policy="max", height=400,output_backend='webgl')


xaxis = LinearAxis(axis_label='HEY IM HERE')
yaxis = LinearAxis(axis_label='HEY IM HERE')
xgrid = Grid()
ygrid= Grid(ticker=yaxis.ticker)
p.add_layout(xaxis,'below')
p.add_layout(yaxis,'left')
p.add_layout(xgrid,'center')
p.add_layout(ygrid,'center')

cjs = '''
function toSuperscript(num) {
    return String(num).replace(/-/g, '⁻').replace(/\d/g, d => "⁰¹²³⁴⁵⁶⁷⁸⁹"[d]);
}   
return `10${toSuperscript(tick)}`
'''
p.xaxis.ticker = LogTicker()
p.xaxis.formatter = CustomJSTickFormatter(code=cjs)
g = Scatter(x='x',y='y')
p.add_glyph(cds,g)

pan = PanTool()
boxzoom = BoxZoomTool()
wz = WheelZoomTool()
box_select = BoxSelectTool()
lasoo = LassoSelectTool()
reset = ResetTool()
examine = ExamineTool()
tools = [pan, boxzoom, box_select, lasoo, examine,
         wz, reset]
p.add_tools(*tools)
p.toolbar.active_scroll = wz
p.toolbar.autohide = True




output_notebook(hide_banner=True)

show(p)
      
```

```python
from bokeh.plotting import figure
# generate random data from a powerlaw
measured = 1/np.random.power(2.5, 100000)

hist, edges = np.histogram(measured, bins=50)

p = figure(title="Power law (a=2.5)", x_axis_type="log", y_axis_type='log')
p.quad(top=hist, bottom=0.001, left=edges[:-1], right=edges[1:], line_color=None)
show(p)
```
